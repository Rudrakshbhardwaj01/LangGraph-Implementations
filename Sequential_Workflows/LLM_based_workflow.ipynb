```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
 
```

```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA

model = ChatNVIDIA(
  model="openai/gpt-oss-20b",
  api_key="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", 
  temperature=1,
  top_p=1,
  max_tokens=4096,
)

```

```python
class LLMstate(TypedDict):
    question:str
    answer:str
```

```python
def LLMQA(state:LLMstate)->LLMstate:
    question=state["question"]
    prompt=f"Answer the following question {question}"
    answer=model.invoke(prompt).content

    state["answer"]=answer

    return state
```

```python
graph=StateGraph(LLMstate)

graph.add_node("LLMQA",LLMQA)

graph.add_edge(START,"LLMQA")
graph.add_edge("LLMQA",END)

workflow=graph.compile()
```

```python
initial_state={"question":"How far is the moon from the earth?"}
final_state=workflow.invoke(initial_state)
print(final_state["answer"])
```
