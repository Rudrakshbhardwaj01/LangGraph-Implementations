```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
 
```

```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA

model = ChatNVIDIA(
  model="openai/gpt-oss-20b",
  api_key="nvapi-Key", 
  temperature=1,
  top_p=1,
  max_tokens=4096,
)

```

```python
class BlogState(TypedDict):
    title:str
    outline:str
    content:str
    score:int
```

```python
def create_outline(state:BlogState)->BlogState:
    title=state["title"]
    prompt=f"Generate a detailed outline of the topic-{title} to create a blog"
    outline=model.invoke(prompt).content

    state["outline"]=outline

    return state

def create_blog(state:BlogState)->BlogState:
    title=state["title"]
    outline=state["outline"]
    prompt=f"Generate a detailed blog on the topic- {title} using {outline} as the context"
    content=model.invoke(prompt).content

    state["content"]=content

    return state

def generate_score(state:BlogState)->BlogState:
    title=state["title"]
    outline=state["outline"]
    content=state["content"]
    prompt=f"Rate the following content-{content} keeping this outline-{outline} as the rating standard/benchmark and give a score out of 10. Remember, you only have to generate a score out of 10, the response of this prompt should just be just an integer."
    score=model.invoke(prompt).content

    state["score"]=score

    return state
```

```python
graph=StateGraph(BlogState)

graph.add_node("create_outline",create_outline)
graph.add_node("create_blog",create_blog)
graph.add_node("generate_score",generate_score)

graph.add_edge(START,"create_outline")
graph.add_edge("create_outline","create_blog")
graph.add_edge("create_blog","generate_score")
graph.add_edge("generate_score",END)

workflow=graph.compile()
```

```python
initial_state={"title":"Rise of AI in India"}

final_state=workflow.invoke(initial_state)
```
